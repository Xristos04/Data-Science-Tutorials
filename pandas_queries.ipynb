```python

# Input: 
# Customers table:
# +----+-------+
# | id | name  |
# +----+-------+
# | 1  | Joe   |
# | 2  | Henry |
# | 3  | Sam   |
# | 4  | Max   |
# +----+-------+
# Orders table:
# +----+------------+
# | id | customerId |
# +----+------------+
# | 1  | 3          |
# | 2  | 1          |
# +----+------------+
# Output: 
# +-----------+
# | Customers |
# +-----------+
# | Henry     |
# | Max       |
# +-----------+
customers_data = {'id': [1, 2, 3, 4],
                  'name': ['Joe', 'Henry', 'Sam', 'Max']}

orders_data = {'id': [1, 2],
               'customerId': [3, 1]}

customer_df = pd.DataFrame(customers_data)
order_df = pd.DataFrame(orders_data)

def customers_who_never_order():

     merged_df =  pd.merge(customer_df, order_df, left_on='id',right_on='customerId', how='left')
     return merged_df[merged_df['customerId'].isnull()]['name']

output = customers_who_never_order()
output
```

```python

# Input: 
# Employee table:
# +----+-------+--------+-----------+
# | id | name  | salary | managerId |
# +----+-------+--------+-----------+
# | 1  | Joe   | 70000  | 3         |
# | 2  | Henry | 80000  | 4         |
# | 3  | Sam   | 60000  | Null      |
# | 4  | Max   | 90000  | Null      |
# +----+-------+--------+-----------+

data = [[1, 'Joe', 70000, 3], [2, 'Henry', 80000, 4], [3, 'Sam', 60000, None], [4, 'Max', 90000, None]]
employee = pd.DataFrame(data, columns=['id', 'name', 'salary', 'managerId']).astype({'managerId' :'Int64'})

def employee_earning_more_than_their_managers(employee : pd.DataFrame):

    v = employee.merge(employee, how='inner', left_on='managerId', right_on='id')
    v = v[v['salary_x']> v['salary_y']][['name_x']]
    return v

output = employee_earning_more_than_their_managers(employee)
output
```

```python
# Person table:
# +----------+----------+-----------+
# | personId | lastName | firstName |
# +----------+----------+-----------+
# | 1        | Wang     | Allen     |
# | 2        | Alice    | Bob       |
# +----------+----------+-----------+

# Address table:
# +-----------+----------+---------------+------------+
# | addressId | personId | city          | state      |
# +-----------+----------+---------------+------------+
# | 1         | 2        | New York City | New York   |
# | 2         | 3        | Leetcode      | California |
# +-----------+----------+---------------+------------+


# Output: 
# +-----------+----------+---------------+----------+
# | firstName | lastName | city          | state    |
# +-----------+----------+---------------+----------+
# | Allen     | Wang     | Null          | Null     |
# | Bob       | Alice    | New York City | New York |
# +-----------+----------+---------------+----------+

data = [[1, 'Allen', 'Wang'], [2, 'Bob', 'Alice']]
person = pd.DataFrame(data, columns=['personId', 'firstName', 'lastName']).astype({'personId':'Int64', 'firstName':'object', 'lastName':'object'})
data = [[1, 2, 'New York City', 'New York'], [2, 3, 'Leetcode', 'California']]
address = pd.DataFrame(data, columns=['addressId', 'personId', 'city', 'state']).astype({'addressId':'Int64', 'personId':'Int64', 'city':'object', 'state':'object'})

def get_details():

    data = pd.merge(person, address, how='left', left_on='personId', right_on='personId')
    data = data[['firstName','lastName','city', 'state']]
    return data

output = get_details()
output
```

```python
# TODO
# 177. Nth Highest Salary

# Input: 
# Employee table:
# +----+--------+
# | id | salary |
# +----+--------+
# | 1  | 100    |
# | 2  | 200    |
# | 3  | 300    |
# +----+--------+
# n = 2
# Output: 
# +------------------------+
# | getNthHighestSalary(2) |
# +------------------------+
# | 200                    |
# +------------------------+
```

```python
data = [[3,300],[1, 100],[2,200]]
employee = pd.DataFrame(data, columns=['id','salary'])


def n_highest_sal(n_value):

    if n_value > len(employee):
        n_highest = None
       
    else:
        n_value = n_value -1

        sorted_employed = employee['salary'].sort_values(ascending=True)

        n_highest = sorted_employed.iloc[n_value]

    return pd.DataFrame([n_highest], columns=["salary"])


n_highest_sal(7)
```

```python
# TODO
# 178. Rank Scores
# Input: 
# Scores table:
# +----+-------+
# | id | score |
# +----+-------+
# | 1  | 3.50  |
# | 2  | 3.65  |
# | 3  | 4.00  |
# | 4  | 3.85  |
# | 5  | 4.00  |
# | 6  | 3.65  |
# +----+-------+
# Output: 
# +-------+------+
# | score | rank |
# +-------+------+
# | 4.00  | 1    |
# | 4.00  | 1    |
# | 3.85  | 2    |
# | 3.65  | 3    |
# | 3.65  | 3    |
# | 3.50  | 4    |
# +-------+------+
```

```python
data = [[1 , 3.5],[2 , 3.65],[3 , 4],[4 , 3.85],[5 , 4],[6 , 3.65]]
scores = pd.DataFrame(data, columns=['id','score'])

sorted_scores = scores.sort_values(by=["score"], ascending=False)
rankes_sc = sorted_scores.rename(columns={"id":"rank"})
rankes_sc["rank"] = rankes_sc["score"].rank(method="dense", ascending=False)
rankes_sc
```
