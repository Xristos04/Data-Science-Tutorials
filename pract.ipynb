```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

# Δημιουργία συνθετικών δεδομένων για παράδειγμα
data = {'Rating': np.random.normal(2, 1, 40),
        'Year': np.random.randint(2010, 2021, 40)}

df = pd.DataFrame(data)
df
```

```python
train_data, test_data = train_test_split(df, test_size=0.2, random_state=42)

# Διαχωρισμός των χαρακτηριστικών (X) και της μεταβλητής κατάστασης (y)
X_train = train_data[['Year']]
y_train = train_data['Rating']

X_test = test_data[['Year']]
y_test = test_data['Rating']

# Εκπαίδευση του μοντέλου γραμμικής παλινδρόμησης
model = LinearRegression()
model.fit(X_train, y_train)

# Πρόβλεψη των τιμών για το 2022
predictions_2022 = model.predict([[2022]])

# Εκτύπωση των προβλεπόμενων τιμών
print("predict for 2022:", predictions_2022[0])
```

```python
# Δημιουργία τυχαίων δεδομένων
np.random.seed(42)
X = np.random.rand(100, 1) * 10  # Χ ως τυχαία αριθμητική σειρά
y = 2 * X + 1 + np.random.randn(100, 1) * 2  # Υ = 2Χ + 1 + θόρυβος

# Δημιουργία DataFrame
df = pd.DataFrame({'X': X.flatten(), 'y': y.flatten()})

df
```

```python
# Εκτύπωση γραφήματος των δεδομένων
plt.scatter(X, y, label='Πραγματικά δεδομένα')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()

# Εφαρμογή γραμμικής παλινδρόμησης
model = LinearRegression()
model.fit(X, y)

# Εκτύπωση των συντελεστών
print('Συντελεστής πλεονάζοντος:', model.intercept_[0])
print('Συντελεστής κλίσης (βάρος):', model.coef_[0][0])

# Πρόβλεψη για νέα δεδομένα
new_data = np.array([[3], [6]])
predictions = model.predict(new_data)
print('Προβλεπόμενες τιμές για τα νέα δεδομένα:', predictions.flatten())

# Εκτύπωση της γραφικής παράστασης
plt.scatter(X, y, label='Πραγματικά δεδομένα')
plt.plot(X, model.predict(X), color='red', linewidth=3, label='Προβλεπόμενα δεδομένα')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```
