## LINE PLOT

```python
import numpy as np

import matplotlib.pyplot as plt
#Define my data

x = np.arange(0,10,0.1)
y = 2*x**2 + x + 5
plt.plot(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Equation")
plt.show()
```

## Bar Plot

```python
import matplotlib.pyplot as plt

data = {'apples':20 , 'banana':10, "cherry":5 }
names = list(data.keys())
values = list(data.values())

fig =  plt.figure(figsize=(10,5))

plt.bar(names, values)
plt.show()
```

## Horizontal

```python
import matplotlib.pyplot as plt

data = {'apples':20 , 'banana':10, "cherry":5 }
names = list(data.keys())
values = list(data.values())

fig =  plt.figure(figsize=(10,5))

plt.barh(names, values)
plt.title("Horizontal Bar")
plt.show()
```

## Histogram

```python
import matplotlib.pyplot as plt

numbers = [10, 12, 13, 45, 67, 23, 45, 89, 12, 45, 90, 32]
# CHECK IT OUT
# number = np.asarray(numbers)

# OPTION : cumulative=True
plt.hist(numbers, bins=[0,20,40,60,80,100], edgecolor='#FEDEFE', color='#FF2332')
plt.title("Histogram Plot Demo")
plt.xlabel("Range of values")
plt.ylabel("Frequencies")
plt.grid(True)
plt.show()
```

## Advanced Plots

```python
import matplotlib.pyplot as plt

# Data Preparation
total = [20,4,1,34,43,22,3,3,2,23,4]
order = [2,2,1,3,14,2,18,7,9,21,9]
discount = [30,10,20,40,40,10,10,5,40,20,60]

data = list([total,order, discount])

plt.boxplot(data, showmeans=True)
plt.title("Box Plot")
plt.xlabel("Range of values")
plt.ylabel("Frequencies")
plt.grid(True)
plt.show()
```

## Violin Plot

```python
import matplotlib.pyplot as plt

def violin_plot():
    # Data Preparation
    total = [20,4,1,34,43,22,3,3,2,23,4]
    order = [2,2,1,3,14,2,18,7,9,21,9]
    discount = [30,10,20,40,40,10,10,5,40,20,60]

    data = list([total,order, discount])

    plt.violinplot(data, showmeans=True)
    plt.title("Violin Plot")
    plt.xlabel("Range of values")
    plt.ylabel("Frequencies")
    plt.grid(True)
    plt.show()
```

```python
violin_plot()
```

```python
import matplotlib.pyplot as plt

# Data Preparation
total = [20,4,1]
labels = ["Guitars", "Violins", "Drums"]


plt.pie(total, labels=labels, autopct='%1.2f%%' ,shadow= True, startangle=90)
plt.title("Pie Plot")
plt.xlabel("Range of values")
plt.ylabel("Frequencies")
plt.grid(True)
plt.show()
```
